```python
def get_word_score(word, letter_values):
    total = 0
    for letter in word:
        if letter in letter_values:
            total += letter_values[letter]
    return total


def value():
    letter_values = {
        'a':1, 'b':3, 'c':3, 'd':2, 'e':1,
        'f':4, 'g':2, 'h':4, 'i':1, 'j':8,
        'k':5, 'l':1, 'm':3, 'n':1, 'o':1,
        'p':3, 'q':10, 'r':1, 's':1, 't':1,
        'u':1, 'v':4, 'w':4, 'x':8, 'y':4, 'z':10
    }

    while True:
        word = input("Enter a word (or 'q' to quit): ").lower()

        if word == "q" or word == "quit":
            print("Done")
            break

        score = get_word_score(word, letter_values)
        print("Score:", score)


if __name__ == "__main__":
    value()
```
